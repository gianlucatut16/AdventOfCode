```python
def check(pos_init, matrix):
    count = 0
    for i in (-1, 0, 1):
        for j in (-1, 0, 1):
            if i == 0 and j == 0:
                continue
            new_pos = (pos_init[0] + i, pos_init[1] + j)
            if new_pos[0] >= 0 and new_pos[1] >= 0 and new_pos[0] < len(matrix[0]) and new_pos[1] < len(matrix[0]):
                if matrix[new_pos[0]][new_pos[1]] == '@':
                    count+= 1
            else:
                continue

    if count < 4:
        return True
    else:
        return False                
            
```

```python
with open('input.txt', 'r') as f: 
    matrix_lst = []
    for line in f:
        line_str = line.strip()
        matrix_lst.append([x for x in line_str])

    second_count = 0
    first_count = 1
    ind=0

    while first_count != 0:
        first_count = 0
        rimuovere = []
        for i in range(len(matrix_lst)):
            for j in range(len(matrix_lst[0])):
                if matrix_lst[i][j] == '@':
                    if check((i,j), matrix_lst):
                        first_count += 1
                        rimuovere.append((i, j))

        second_count += first_count
        
        if ind == 0:
            print(f'Soluzione per la prima parte: {first_count}')
            ind += 1

        if first_count > 0:
            for tup in rimuovere:
                matrix_lst[tup[0]][tup[1]] = '.'

               

print(f'Soluzione per la seconda parte: {second_count}')
```
