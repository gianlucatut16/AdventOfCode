```python
def beam(dict_beam, f, counter):
    line = f.readline()
    line = f.readline()
    
    if '^' not in line:
        return dict_beam, counter
    
    else:
        dict_beam_new = {}
        for i in dict_beam.keys():
            if line[i] == '^':
                counter += 1

                if (i - 1) not in dict_beam_new.keys():
                    dict_beam_new[i-1] = dict_beam[i]
                else:
                    dict_beam_new[i-1] += dict_beam[i]

                if (i+1) not in dict_beam_new.keys():
                    dict_beam_new[i+1] = dict_beam[i]
                else:
                    dict_beam_new[i+1] +=  dict_beam[i]
            else:
                if i not in dict_beam_new.keys():
                    dict_beam_new[i] = dict_beam[i]
                else:
                    dict_beam_new[i] += dict_beam[i]

        return beam(dict_beam_new, f, counter = counter)

    


with open('input.txt', 'r') as f: 
    
    line = f.readline().strip()
    dict_beam = {line.index('S'): 1}
    dict_def, first_count = beam(dict_beam, f, counter = 0)

second_count = 0
for i in dict_def.values():
    second_count += i

print(f'Soluzione per la prima parte: {first_count}')
print(f'Soluzione per la seconda parte: {second_count}')
```
