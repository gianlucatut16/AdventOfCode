```python
def validness(num):
    num_str = str(num)
    if len(num_str) % 2 == 0:
        half = len(num_str) // 2
        if num_str[: half] == num_str[half:]:
            return False
    return True

def validness2(num):
    num_str = str(num)
    half = len(num_str) // 2
    for ind in range(1, half + 1):
        chunk = num_str[:ind]
        if num_str == chunk * (len(num_str)// ind):
            return False
    return True
```

```python
with open('input.txt', 'r') as f:    
    input_file = f.readline()
    list_range = input_file.split(',')
    result1, result2 = 0, 0
    for num_range in list_range:
        num1, num2 = [int(x) for x in num_range.split('-')]
        for num_to_check in range(num1, num2 + 1):
            if validness(num_to_check) == False:
                result1 += num_to_check
            if validness2(num_to_check) == False:
                result2 += num_to_check


print(f'Soluzione per la prima parte: {result1}')
print(f'Soluzione per la seconda parte: {result2}')

```
