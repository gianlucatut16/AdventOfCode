```python
def check_num(limits, num):
    check = False
    for i in range(len(limits)):
        if num >= limits[i][0] and num <= limits[i][1]:
            check = True
            break
    
    return check
    
with open('input.txt', 'r') as f: 
    check = 0
    first_count = 0
    limits = []

    for line in f:
        line_str = line.strip()

        if line_str == '':
            check = 1
            continue

        if check == 0:
            r1, r2 = [int(x) for x in line_str.split('-')]
            if (r1, r2) not in limits:
                limits.append((r1, r2))
            
            else:
                continue

        elif check == 1:
             num = int(line_str)
             if check_num(limits, num):
                 first_count += 1

print(f'Soluzione per la prima parte: {first_count}')
```

```python
def group_lst(lista):
    limits_un = [lista[0]]
    while lista != limits_un:
        for i in range(1, len(lista)):
            tup_ac = lista[i]
            check = 0
            for j in range(len(limits_un)):
                tup_ck = limits_un[j]
                if (tup_ck[0] <= tup_ac[0] <= tup_ck[1]) and tup_ac[1] >= tup_ck[1]:
                    limits_un[j] = tup_ck[0], tup_ac[1]
                    check = 1
                    break
                elif (tup_ck[0] <= tup_ac[1] <= tup_ck[1]) and tup_ac[0] <= tup_ck[0]:
                    limits_un[j] = tup_ac[0], tup_ck[1]
                    check = 1
                    break
                elif tup_ac[0] <= tup_ck[0] and tup_ac[1] >= tup_ck[1]:
                    limits_un[j] = tup_ac
                    check = 1
                    break
                elif tup_ac[0] >= tup_ck[0] and tup_ac[1] <= tup_ck[1]:
                    limits_un[j] = tup_ck
                    check = 1
                    break
            if check == 0:
                limits_un.append(tup_ac)
        
        if lista == limits_un:
            break
        else:
            lista = limits_un
            limits_un = lista[0]
            if type(limits_un) == tuple:
                limits_un = [limits_un]
    return limits_un
        
result = group_lst(limits)

second_count = 0
for tup in result:
    second_count += (tup[1] - tup[0]) + 1
print(f'Soluzione per la seconda parte: {second_count}')
            
```
