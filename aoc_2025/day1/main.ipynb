```python
with open('input.txt', 'r') as f:    
    first_count, second_count = 0, 0
    punct = 50
    for line in f:
        move = line[0]
        steps = int(line[1:])
        second_count += (steps // 100)    # Commenta per avere solo la prima parte
        true_steps = steps % 100
        if move == 'R':
            punct += true_steps  

            if punct > 100:
                punct -= 100
                second_count += 1                # Commenta per avere solo la prima parte
                
            elif punct == 100:
                punct = 0    
        else:
            if punct == 0:
                second_count -= 1                # Commenta per avere solo la prima parte

            punct -= true_steps 
            if punct < 0:
                second_count += 1                # Commenta per avere solo la prima parte
                punct = 100 - abs(punct)
    
        if punct in (0, 100):
            second_count += 1
            first_count += 1

print(f'Soluzione per la prima parte: {first_count}')
print(f'Soluzione per la seconda parte: {second_count}')
```
