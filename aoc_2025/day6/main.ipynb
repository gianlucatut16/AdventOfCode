```python
def first_sum(list_numbers):
    first_count = 0
    for i in range(len(list_numbers[0])):
        
        if list_numbers[-1][i] == '*':
            part_result = 1
            for j in range(len(list_numbers) - 1):
                part_result *= int(list_numbers[j][i])
        else:
            part_result = 0
            for j in range(len(list_numbers) - 1):
                part_result += int(list_numbers[j][i])

        first_count += part_result

    return first_count
    
def do_math(numbers, op):
    if op == '+':
        partial_result = 0
        for num in numbers:
            partial_result += num
    elif op == '*':
        partial_result = 1
        for num in numbers:
            partial_result *= num

    return partial_result
```

```python
with open('input.txt', 'r') as f: 
    check = 0
    first_count = 0
    limits = []

    numbers_operations, list_str = [], []
    for line in f:
        line_str = line.strip()
        list_str.append(line.replace('\n', ''))


        numbers_operations.append(([x for x in line_str.split(' ') if x != '']))

    first_count = first_sum(numbers_operations)

    print(f'Soluzione per la prima parte: {first_count}')


    second_count = 0
    numbers = []


    for i in range(1, len(list_str[0]) + 1):
        numero = ''
        for j in range(len(list_str) - 1):
            if list_str[j][-i] != '':
                numero += list_str[j][-i]
            else: 
                continue
        try:
            numbers.append(int(numero))
        except ValueError:
            continue
        

        if list_str[j + 1][-i] in ('+', '*'):
            op = list_str[j + 1][-i]
            second_count += do_math(numbers, op)
            numbers = []

    print(f'Soluzione per la seconda parte: {second_count}')
```
