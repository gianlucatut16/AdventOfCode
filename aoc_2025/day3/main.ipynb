```python
def get_next_number(text, remaining, num_batt):

    if len(text) == num_batt:
        return text
    
    else:
        possible_numbers = [int(i) for i in remaining]
        check = False
        ind_grandezza = -1
        while check == False:
            first_number = str(sorted(possible_numbers)[ind_grandezza])
            ind = remaining.index(first_number)
            remaining_tmp = remaining[ind + 1:]
            if len(remaining_tmp) < (num_batt-1) - len(text):
                ind_grandezza -= 1
            else:
                remaining = remaining_tmp
                check = True
        if len(remaining) >= (num_batt-1) - len(text):
            text += first_number
            return get_next_number(text, remaining, num_batt)
        elif len(remaining) == (num_batt-1) - len(text):
            text += remaining[:]
            return text
        else:
            return False
        

def get_batteries_rec(text, num_batt):
    max_result = 0

    for ind in range(0, len(text) - (num_batt - 1)):
        first_chunk = text[ind]
        remaining = text[ind + 1:]
        max_tmp = get_next_number(first_chunk, remaining=remaining, num_batt = num_batt)
        if max_tmp == False:
            continue
        else:
            max_tmp = int(max_tmp)
            if max_tmp > max_result:
                max_result = max_tmp
    return max_result
```

```python
with open('input.txt', 'r') as f:    
    first_count, second_count = 0, 0
    for line in f:
        first_count += get_batteries_rec(line.strip(), 2)
        second_count += get_batteries_rec(line.strip(), 12)
        

print(f'Soluzione per la prima parte: {first_count}')
print(f'Soluzione per la seconda parte: {second_count}')

```
