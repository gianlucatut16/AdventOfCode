```python
import numpy

with open('real_input.txt', 'r') as file:
    data = file.read()


directions, updates = data.split('\n\n')
directions = directions.split('\n')
prec_nums = [int(item[:2]) for item in directions]
succ_nums = [int(item[3:]) for item in directions]
```

```python
before_dict = {}

for num in prec_nums:
    tmp = []
    for nest_ind in range(len(succ_nums)):
        if prec_nums[nest_ind] == num:
            tmp.append(succ_nums[nest_ind])

    before_dict[num] = tmp




after_dict = {}

for num in succ_nums:
    tmp = []
    for nest_ind in range(len(prec_nums)):
        if succ_nums[nest_ind] == num:
            tmp.append(prec_nums[nest_ind])

    after_dict[num] = tmp

print(before_dict)
print(after_dict)
```

```python
updates_1 = updates.split('\n')
updates_lst = []
for string in updates_1:
    it = [int(_) for _ in string.split(',')]
    updates_lst.append(it)
```

```python
def check_update(update):
    check = True
    for j in range(len(update)):
        

        num = update[j]
        for _ in update[j + 1:]:
            if num in after_dict.keys() and _ in after_dict[num]:
                check = False
                break  

    if check == True:
        central = update[int((len(update) - 1)/2)]
        return True, central
    else:
        return False, None
```

```python
sum = 0
incorrect_updates = []
for update in updates_lst:
    check, central = check_update(update)

    if check == True:
        sum += update[int((len(update) - 1)/2)]
    else:
        if update not in incorrect_updates:
            incorrect_updates.append(update)

print(sum)
```

```python
correct_updates = []
sum_2 = 0
for ind in range(len(incorrect_updates)):
    update = incorrect_updates[ind][:]
    i = 0
    while i < len(update) - 1:
        num = update[i]
        error = False
        for j in range(i + 1, len(update)):
            if num in after_dict.keys() and update[j] in after_dict[num]:
                update[i], update[j] = update[j], update[i]
                i= 0
                error = True
                break
        if error == True:
            i = 0
        else:
            i += 1
    sum_2 += update[int((len(update) - 1)/2)]

print(sum_2)


    
```
