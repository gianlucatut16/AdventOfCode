```python
def check_coord_in_grid(row, col, grid):
    if (0 <= row < len(grid)) and (0 <= col < len(grid[0])):
        return True
    else:
        return False
    


def find_trail(row_ind, col_ind):
    paths = [[(row_ind, col_ind)]]
    while len(paths[-1]) != 10:
        
        new_paths = []
        for path in paths:
            poss_path = []
            row_ind, col_ind = path[-1][0], path[-1][1]

            if check_coord_in_grid(row_ind + 1, col_ind, grid) and (grid[row_ind + 1][col_ind] == len(path)):
                poss_path.append(path.copy() + [(row_ind+1, col_ind)])

            if check_coord_in_grid(row_ind - 1, col_ind, grid) and (grid[row_ind - 1][col_ind] == len(path)):
                poss_path.append(path.copy() + [(row_ind - 1, col_ind)])

            if check_coord_in_grid(row_ind, col_ind + 1, grid) and (grid[row_ind][col_ind + 1] == len(path)):
                poss_path.append(path.copy() + [(row_ind, col_ind + 1)])

            if check_coord_in_grid(row_ind, col_ind - 1, grid) and (grid[row_ind][col_ind - 1] == len(path)):
                poss_path.append(path.copy() + [(row_ind, col_ind - 1)])
            new_paths += poss_path
        paths = new_paths

    
    return paths

def count_score(paths):
    if len(paths) == 1:
        return 1
    else:
        distinct_nine = []
        for item in paths:
            if item[-1] not in distinct_nine:
                distinct_nine.append(item[-1])
        return len(distinct_nine)





with open('real_input.txt', 'r') as file:
    grid = []
    for line in file:
        line = line.strip()
        line_num = [int(item) for item in line]
        grid.append(line_num)


    
result, result_2 = 0, 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == 0:
            paths = find_trail(i,j)
            score = count_score(paths)
            result += score
            result_2 += len(paths)

print(f'Part 1: {result}')
print(f'Part 2: {result_2}')
```
