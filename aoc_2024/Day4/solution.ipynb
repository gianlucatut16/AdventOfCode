```python
import numpy as np

groups = list(np.loadtxt('real_input.txt', dtype = str))

# Horizontal
score_h = 0
for row_ind in range(len(groups)):
    group = groups[row_ind]
    # print(group)
    for col_ind in range(len(groups[0])):
        if group[col_ind] == 'X':
            if group[col_ind: col_ind + 4] == 'XMAS':
                score_h += 1
            if group[col_ind - 3: col_ind + 1] == 'SAMX':
                score_h += 1


# Vertical
score_v = 0
for col_ind in range(len(groups[0])):
    group = [groups[i][col_ind] for i in range(len(groups))]
    group = ''.join(group)
    for row_ind in range(len(group)):
        if group[row_ind] == 'X':
            if group[row_ind: row_ind + 4] == 'XMAS':
                score_v += 1
            if group[row_ind - 3: row_ind] == 'SAM':
                score_v += 1


# Diagonal

score_diag = 0
for row_ind in range(len(groups)):
    for col_ind in range(len(groups[0])):
        if groups[row_ind][col_ind] == 'X':
            # Top-right
            top_right = ''
            bot_right = ''
            top_left = ''
            bot_left = ''

            # Top-right
            if col_ind + 3 < len(groups[0]) and row_ind - 3 >= 0:
                for j in range(1,4):
                    top_right += groups[row_ind - j][col_ind + j]
                    if top_right == 'MAS':
                        score_diag += 1

            # Bottom-right
            if col_ind + 3 < len(groups[0]) and row_ind + 3 < len(groups):
                for j in range(1,4):
                    bot_right += groups[row_ind + j][col_ind + j]
                    if bot_right == 'MAS':
                        score_diag += 1
            
            # Top-left
            if col_ind - 3 >= 0 and row_ind - 3 >= 0:
                for j in range(1,4):
                    top_left += groups[row_ind - j][col_ind - j]
                    if top_left == 'MAS':
                        score_diag += 1

            # Bottom-left
            if col_ind - 3 >= 0 and row_ind + 3 < len(groups):
                for j in range(1,4):
                    bot_left += groups[row_ind + j][col_ind - j]
                    if bot_left == 'MAS':
                        score_diag += 1



print(score_v, score_h, score_diag)
print(f'Result: {score_v + score_h + score_diag}')
```

# Part-two

```python
score = 0
for row_ind in range(len(groups)):
    for col_ind in range(len(groups[0])):
        if row_ind == 0 or row_ind == len(groups)-1 or col_ind == 0 or col_ind == len(groups[0]) - 1:
            continue
    
        if groups[row_ind][col_ind] == 'A':
        
            first_d = groups[row_ind - 1][col_ind - 1] + groups[row_ind][col_ind] + groups[row_ind + 1][col_ind + 1]
            second_d = groups[row_ind - 1][col_ind + 1] + groups[row_ind][col_ind] + groups[row_ind + 1][col_ind - 1]
            if (first_d == 'MAS' or first_d == 'SAM') and (second_d == 'MAS' or second_d == 'SAM'):
                score += 1                

print(f'Result: {score}')
```
