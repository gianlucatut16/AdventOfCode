```python
import numpy as np

input = np.loadtxt('real_input.txt', dtype = int)

distances_lst = []

left_lst = input[:, 0]
right_lst = input[:, 1]

srt_left = sorted(left_lst)
srt_right = sorted(right_lst)

for i in range(len(srt_left)):
    diff = abs(srt_left[i] - srt_right[i])
    distances_lst.append(diff)

print(np.sum(distances_lst))
```

```python
similarity_lst = []

for num_1 in left_lst:
    score = 0
    for num_2 in right_lst:
        if num_2 == num_1:
            score += 1
    similarity_lst.append(num_1 * score)

print(np.sum(similarity_lst))
```
