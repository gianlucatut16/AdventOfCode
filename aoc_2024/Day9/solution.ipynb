```python
with open('real_input.txt', 'r') as file:
    input = file.readline()


def string_to_dict(string):
    output_dict = {}
    id_count = 0
    for ind in range(len(string)):
        value = string[ind]
        if value != '0':
            if ind % 2 == 0:
                output_dict[id_count] = int(value)
                id_count += 1
            else:
                output_dict[str(id_count)] = int(value)
        else:
            continue


    return output_dict




def dict_to_string(input_dict):
    str_output = ''
    for item in input_dict:
        
        value = input_dict[item]
        if type(item) == int:
            str_output += str(item) * value
            # str_output += f'{str(item)}/' * value
        elif type(item) == str:
            str_output += '.' * value
            # str_output += './' * value
        # str_output += '/'
    return str_output


dict_string = string_to_dict(input)
print(dict_string)
string_repr = dict_to_string(dict_string)
print(string_repr)

def move_things_up(string_input):
    str_split = string_input.split('/')[:-1]
    # print(str_split)
    for ind in range(len(str_split)):

        if str_split[ind:] == ['.'] * len(str_split[ind:]):
            break
        if str_split[ind] == '.':
            end_ind = -1
            while str_split[end_ind] == '.':
                end_ind -= 1
            
            str_split[ind], str_split[end_ind] = str_split[end_ind], str_split[ind]
            
                
        else:
            continue
        
    return str_split

def checksum(string_lst):
    result = 0
    for ind in range(len(string_lst)):
        value = string_lst[ind]
        if value not in ('.', '/'):
            result += ind * int(string_lst[ind])
        else:
            continue
    return result

string_result = move_things_up(string_repr)
print(string_result)
print(checksum(string_result))
```
