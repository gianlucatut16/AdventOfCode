```python
def second_rule(string):
    left_half, right_half = string[:int(len(string) / 2)], string[int(len(string) / 2):]
    return int(left_half), int(right_half)

memory = {}
def solve(stone, blinks):
    if blinks == 0:
        return 1
    
    elif stone == 0:
        val = solve(1, blinks -1)

    elif (stone, blinks) in memory:
        return memory[(stone, blinks)]
    
    elif len(str(stone)) % 2 == 0:
        num_1, num_2 = second_rule(str(stone))
        val = solve(num_1, blinks - 1) + solve(num_2, blinks - 1) 
    
    else:
        val = solve(stone * 2024, blinks - 1)
    memory[(stone, blinks)] = val
    return val

NUM_BLINKS = 75
input = '5 89749 6061 43 867 1965860 0 206250'
line = [int(item) for item in input.split(' ')]

result = 0
for stone in line:
    result += solve(stone, NUM_BLINKS)



print(f'Part 2: {result}')


```
