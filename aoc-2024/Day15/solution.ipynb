```python
with open('real_input.txt', 'r') as file:
    data = file.read()

grid, moves = data.split('\n\n')[0].split('\n'), data.split('\n\n')[1]

moves = ''.join(moves.split('\n'))



def movement_trad(move):
    trad_dict = {'>': (0, 1), '<': (0, -1), 'v': (1, 0), '^': (-1, 0)}
        
    if move in trad_dict:
        return trad_dict[move]
    else:
        
        print(move ,'Something wrong!!!!')



obstacles = []


for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == '@':
            pos = i,j
        elif grid[i][j] == 'O':
            obstacles.append([i, j])


for move in moves:
    move_tup = movement_trad(move)
    new_pos = pos[0] + move_tup[0], pos[1] + move_tup[1]

    if (grid[new_pos[0]][new_pos[1]] != '#') and ([new_pos[0], new_pos[1]] not in obstacles):
        pos = new_pos

    elif grid[new_pos[0]][new_pos[1]] == '#':
        pos = pos
        continue
        
    
    elif [new_pos[0],new_pos[1]] in obstacles:

        i, j= new_pos[0], new_pos[1]

        if move_tup == (0, 1):
            while [i, j] in obstacles:
                j += 1

            if grid[i][j] == '#':
                continue

            elif [i, j] not in obstacles:
                for col in range(j, new_pos[1] - 1, -1):
                    if [i, col] in obstacles:
                        ind = obstacles.index([i, col])
                        obstacles[ind] = [i, col + 1]
            pos = new_pos

        if move_tup == (0, -1):
            
            while [i, j] in obstacles:
                j -= 1

            if grid[i][j] == '#':
                continue

            elif [i, j] not in obstacles:
                for col in range(j, new_pos[1] + 1):
                    if [i, col] in obstacles:
                        ind = obstacles.index([i, col])
                        obstacles[ind] = [i, col - 1]
            pos = new_pos


        if move_tup == (1, 0):
            while [i, j] in obstacles:
                i += 1

            if grid[i][j] == '#':
                continue

            elif [i, j] not in obstacles:
                for row in range(i, new_pos[0] - 1, -1):
                    if [row, j] in obstacles:
                        ind = obstacles.index([row, j])
                        obstacles[ind] = [row + 1, j]
            pos = new_pos


        if move_tup == (-1, 0):
            while [i, j] in obstacles:
                i -= 1

            

            if grid[i][j] == '#':
                continue

            elif [i, j] not in obstacles:
                for row in range(i, new_pos[0] + 1):
                    if [row, j] in obstacles:
                        ind = obstacles.index([row, j])
                        obstacles[ind] = [row - 1, j]
            pos = new_pos

result = 0
for obs in obstacles:
    result += (obs[0] * 100) + obs[1]

print('Part 1:', result)
```
