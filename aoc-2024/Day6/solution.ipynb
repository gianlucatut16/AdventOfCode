```python
import numpy


def tuple_sum(tup_1, tup_2):
    return tup_1[0] + tup_2[0], tup_1[1] + tup_2[1]

def check_in_grid(coords, rows_num, cols_num):
    if coords[0] < 0 or coords[0] >= rows_num or coords[1] < 0 or coords[1] >= cols_num:
        return False
    else:
        return True

with open('real_input.txt', 'r') as file:
    data = file.readlines()

grid = [item.strip() for item in data]

canc = 9
guard = 1
blank = 0

grid_num = []
for line_ind in range(len(grid)):
    line = grid[line_ind]
    new_line = []
    for char_ind in range(len(line)):
        if line[char_ind] == '.':
            value = blank
        elif line[char_ind] == '#':
            value = canc
        elif line[char_ind] == '^':
            value = guard
            init_line, init_col = line_ind, char_ind
            print(f'Initial position: {line_ind}, {char_ind}')
        
        new_line.append(value)
    grid_num.append(new_line)

coords = (init_line, init_col)
directions = [(-1,0), (0,1), (1, 0), (0, -1)]
i = 0
visited_coords = []
cond = True
tmp = [coords]
while cond == True:
    direct = directions[i % 4]
    new_coord = tuple_sum(coords, direct)

    if check_in_grid(new_coord, len(grid), len(grid[0])):
        if grid_num[new_coord[0]][new_coord[1]] == 9:
            i += 1
            coords = coords
            visited_coords.append(tmp)
            tmp = []
        else:
            coords = new_coord        
            tmp.append(coords)
    else:
        visited_coords.append(tmp)
        cond = False
    

result_lst = []
for row in visited_coords:
    for tup in row:
        if tup not in result_lst:
            result_lst.append(tup)

print(f'Result: {len(result_lst)}')
```

```python
result_2 = 0
count = 0
for tup in result_lst:
    grid_num_1 = grid_num.copy()
    grid_num_1[tup[0]][tup[1]] = 9
    coords = (init_line, init_col)

    i = 0
    visited_coords = []
    cond = True
    tmp = [coords]

    while cond == True:
        direct = directions[i % 4]
        new_coord = tuple_sum(coords, direct)

        if check_in_grid(new_coord, len(grid_num_1), len(grid_num_1[0])):
            if grid_num_1[new_coord[0]][new_coord[1]] == 9:
                i += 1
                coords = coords
                visited_coords.append(tmp)
                tmp = []
            else:
                coords = new_coord        
                tmp.append(coords)
        else:
            visited_coords.append(tmp)
            cond = False

        if len(visited_coords) > 3:
            try:
                if visited_coords[-1][-1] in [item[-1] for item in visited_coords[:-1] if len(item) > 0]:
                    result_2 += 1
                    break
            except IndexError:
                pass
        else:
            continue


    count+=1
    grid_num_1[tup[0]][tup[1]] = 0

print(f'Part two: {result_2}')
```
