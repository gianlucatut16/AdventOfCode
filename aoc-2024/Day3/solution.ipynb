```python
import numpy as np
def check_row(row):
    NUMBERS = '0123456789'
    result= 0

    if row[0] != '(':
        result = 0
    else:
        try:
            # Primo numero
            for i in range(1, 4):
                if row[i] in NUMBERS:
                    comma_ind = i + 1
                else:
                    comma_ind = i
                    break

            if row[1: comma_ind] != '':
                first_num = int(row[1: comma_ind])
            else:
                first_num = 0
            
            # Secondo numero
            for i in range(comma_ind + 1, comma_ind + 4):
                if row[i] in NUMBERS:
                    parent_ind = i + 1
                else:
                    parent_ind = i
                    break
                
            if row[comma_ind + 1: parent_ind] != '':
                second_num = int(row[comma_ind + 1: parent_ind])
            else:
                second_num = 0


            if row[parent_ind] == ')':
        
                result = first_num * second_num
            else:
                result = 0


        except IndexError:
            result

            
    return result
    

```

```python
with open('real_input.txt', 'r') as file:
    solution = 0
    enable = True
    for line in file:

        mults = []
        line=line.strip()
        split_line = line.split('mul')
        split_line = [item for item in split_line if item != '']
        for sent in split_line:
            
            if enable == True:
                mults.append(check_row(sent))
                if "don't()" in sent:

                    enable = False

            
            else:
                mults.append(0)
                if "do()" in sent:
                    enable = True
        
        solution += np.sum(mults)
    print(solution)
```
