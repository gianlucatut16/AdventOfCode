```python
with open('real_input.txt', 'r') as file:
    grid = []
    for line in file:
        line = line.strip()
        grid.append(line)


def token(a_x, a_y, b_x, b_y, prize_x, prize_y):

    det = a_x * b_y - b_x * a_y
    total = 0

    if det == 0:
        max_a = prize_x // a_x
        max_b = prize_x // b_x

        if (a_x * max_a, a_y * max_a) == (prize_x, prize_y) and max_a < 3*max_b:
            total += max_a
        
        elif (b_x * max_b, b_y * max_b) == (prize_x, prize_y):
            total += max_b
    

    else:
        mul_a, mul_b = int((prize_x*b_y - b_x * prize_y)/det), int((a_x* prize_y -prize_x * a_y)/ det)
        if mul_a * a_x + mul_b * b_x == prize_x and mul_a * a_y + mul_b * b_y == prize_y: # if integer solution exists
            total += 3 * mul_a + mul_b


    return total

```

```python
result = 0
params = []
for sent in grid:


    if (sent.split(':')[0] == 'Button A'):
        sent_split = (','.join(sent.split('+'))).split(',')
        a_x, a_y = int(sent_split[1]), int(sent_split[3])
        params += [a_x, a_y]

    elif (sent.split(':')[0] == 'Button B'):
        sent_split = (','.join(sent.split('+'))).split(',')
        b_x, b_y = int(sent_split[1]), int(sent_split[3])
        params += [b_x, b_y]

    elif (sent.split(':')[0] == 'Prize'):
        sent_split = (','.join(sent.split('='))).split(',')
        prize_x, prize_y = int(sent_split[1]) + 10000000000000, int(sent_split[3]) + 10000000000000
        params += [prize_x, prize_y]

    elif sent == '':
        continue

    if len(params) == 6:
        result += token(*params)
        params = []
    
    

result


```
